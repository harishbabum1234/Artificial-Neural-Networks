```python
X=balanced_dff.drop('DEATH_EVENT', axis=1)
y=balanced_dff['DEATH_EVENT']
```

```python
# SPlitting the data
X_train,X_test,y_train,y_test =train_test_split(X,y,test_size=0.33,random_state=42)
print(X_train.shape,y_train.shape)
print(X_test.shape,y_test.shape)
```

```python
# standard scaler mmethod
from sklearn.preprocessing import StandardScaler
scc=StandardScaler()
X_train=scc.fit_transform(X_train)
X_test = scc.transform(X_test)
```

```python
# Sequential model development
classifierss=Sequential()
```

```python
classifierss.add(Dense(units=6,activation='relu',kernel_initializer='he_normal',input_dim = 12))
classifierss.add(Dense(units=5,activation='relu',kernel_initializer='he_normal'))
classifierss.add(Dense(units=1,kernel_initializer='glorot_uniform',activation='sigmoid'))
```

```python
classifierss.compile(optimizer='adam',loss=tf.keras.losses.BinaryCrossentropy(),metrics=tf.keras.metrics.BinaryAccuracy())
```

```python
model_history = classifierss.fit(X_train,y_train,batch_size=10,validation_split=0.30,epochs=50)
```

```python
print(model_history.history.keys())
```

```python
# summarize history for accuracy
plt.plot(model_history.history['binary_accuracy'])
plt.plot(model_history.history['val_binary_accuracy'])

plt.title('Accuracy of model')
plt.ylabel('accuracy')
plt.xlabel('Epoch')
plt.legend(['Train data','Test data'], loc='upper left')
plt.show()
```

```python
# summarization of history for evaluating the accuracy
plt.plot(model_history.history['loss'])
plt.plot(model_history.history['val_loss'])

plt.title('model_accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch ')
plt.legend(['training_data','testing_data'], loc='upper left')
plt.show()
```

```python
y_pred= classifierss.predict(X_test)
y_pred = y_pred>0.5
```

```python
# Model evaluation with CM(confusion matrix)
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)

# accuracy evaluation
from sklearn.metrics import accuracy_score
score=accuracy_score(y_pred,y_test)
```

```python
# printing Confusion matrix
print(cm)

print(score)
```

```python
# Classification report representation
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```
